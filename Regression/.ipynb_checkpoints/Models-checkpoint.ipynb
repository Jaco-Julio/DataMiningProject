```python
#Importing Libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.model_selection import KFold, cross_validate
import warnings
warnings.filterwarnings('ignore')
```

```python
# Loading the training sets and test sets
X_train = np.load('/home/jaco/Documentos/Projetos/Data Minning Project/Regression/X_train.npy')
X_test = np.load('/home/jaco/Documentos/Projetos/Data Minning Project/Regression/X_test.npy')
Y_train = np.load('/home/jaco/Documentos/Projetos/Data Minning Project/Regression/Y_train.npy')
Y_test = np.load('/home/jaco/Documentos/Projetos/Data Minning Project/Regression/Y_test.npy')
X_train_norm = np.load('/home/jaco/Documentos/Projetos/Data Minning Project/Regression/X_train_norm.npy')
X_test_norm = np.load('/home/jaco/Documentos/Projetos/Data Minning Project/Regression/X_test_norm.npy')
Y_train_norm = np.load('/home/jaco/Documentos/Projetos/Data Minning Project/Regression/Y_train_norm.npy')
Y_test_norm = np.load('/home/jaco/Documentos/Projetos/Data Minning Project/Regression/Y_test_norm.npy')
X_train_stand = np.load('/home/jaco/Documentos/Projetos/Data Minning Project/Regression/X_train_stand.npy')
X_test_stand = np.load('/home/jaco/Documentos/Projetos/Data Minning Project/Regression/X_test_stand.npy')
Y_train_stand = np.load('/home/jaco/Documentos/Projetos/Data Minning Project/Regression/Y_train_stand.npy')
Y_test_stand = np.load('/home/jaco/Documentos/Projetos/Data Minning Project/Regression/Y_test_stand.npy')
```

```python
Regressors = {
        'Linear Regressor': LinearRegression(),
        'Ridge Regressor': Ridge(),
        'Lasso Regressor': Lasso(),
        'ElasticNet Regressor': ElasticNet(),
        'KNN Regressor': KNeighborsRegressor(),
        'Decision Tree Regressor': DecisionTreeRegressor(),
        'Suport Vector Regressor': SVR()
}
```

```python
cv = KFold(n_splits=5, random_state=2)
scoring = ['neg_mean_absolute_error','neg_mean_squared_error','r2']
```

```python
for key,Regressor in Regressors.items():
    results = cross_validate(Regressor,X_train,Y_train,cv=cv,scoring=scoring)
    print('---------------------------------------------------------------------------------')
    print('Regresor name: ',key, '\t', 'MAE:',results['test_neg_mean_absolute_error'].mean())
    print('\nRegresor name: ',key, '\t', 'MSE:',results['test_neg_mean_squared_error'].mean())
    print('\nRegresor name: ',key, '\t', 'R2:',results['test_r2'].mean())
    print('---------------------------------------------------------------------------------')
```

```python
for key,Regressor in Regressors.items():
    results = cross_validate(Regressor,X_train_norm,Y_train_norm,cv=cv,scoring=scoring)
    print('---------------------------------------------------------------------------------')
    print('Regresor name: ',key, '\t', 'MAE:',results['test_neg_mean_absolute_error'].mean())
    print('\nRegresor name: ',key, '\t', 'MSE:',results['test_neg_mean_squared_error'].mean())
    print('\nRegresor name: ',key, '\t', 'R2:',results['test_r2'].mean())
    print('---------------------------------------------------------------------------------')
```

```python
for key,Regressor in Regressors.items():
    results = cross_validate(Regressor,X_train_stand,Y_train_stand,cv=cv,scoring=scoring)
    print('---------------------------------------------------------------------------------')
    print('Regresor name: ',key, '\t', 'MAE:',results['test_neg_mean_absolute_error'].mean())
    print('\nRegresor name: ',key, '\t', 'MSE:',results['test_neg_mean_squared_error'].mean())
    print('\nRegresor name: ',key, '\t', 'R2:',results['test_r2'].mean())
    print('---------------------------------------------------------------------------------')
```
